```python
import numpy as np;
import matplotlib.pyplot as plt
```

---

## Perceptron Learning Algorithm 

```python
def PLA(x_val, y_val ,labels, learning_rate=1):
    converged = False
    n=len(x_val)
    updates = 0
    iterations = 0
    b=0
    w=np.zeros(2)

    while not converged:
        converged = True 

        for i in range(n):
            xi = np.array([x_val[i],y_val[i]])
            t = np.dot(w,xi ) + b

            if (labels[i]*(t) <= 0):
                b += learning_rate * labels[i]
                w += learning_rate * labels[i] * xi
                updates += 1
                converged = False

        iterations+=1

    return w, b, updates, iterations


def g(x):
    y = -(w[0] * x + b) / w[1]
    return y

    
```

---
## (a) First Data Set


### Creating First Data set

```python


# Define the target function
def f(x):
    return 2*x + 1

# Generate 20 random data points
np.random.seed(100)  # for reproducibility
x_values = np.random.uniform(-20, 20, 20)
y_values = np.random.uniform(-20, 20, 20)

# Assign labels based on the target function
labels = np.where(y_values > f(x_values), 1, -1)

# Plot the data points
plt.scatter(x_values, y_values, c=labels, cmap=plt.cm.Paired, marker='o', label='Data Points')
# Plot the target function
x_range = np.linspace(-20, 20, 50)
plt.plot(x_range, f(x_range), label='Target Function', color='black')

# Add labels to the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend
plt.legend()

# Show the plot
plt.show()



#> This is a blockquote.


```

```python
w , b , updates , iterations = PLA(x_values, y_values, labels)
print(f"Hypothesis g comes out to be g(x) = {round(-w[0] / w[1] ,2)} x {round(b / w[1],2)} ")

# Plotting the data, target function, and final hypothesis
plt.scatter(x_values, y_values, c=labels, cmap=plt.cm.Paired, marker='o', label='Data Points')
plt.plot(x_range, f(x_range), label='Target Function', color='black')

    
plt.plot(x_range, g(x_range), label='Final Hypothesis (g)', linestyle= 'dashed', color='red')

# Add labels to the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
